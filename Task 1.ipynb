```python
# Import necessary library
  
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.drop(['Series Name','Country Name'],axis=1, inplace=True)
```

```python
# extract  top 15 countries

total_population_data = df[df['Series Code']=='SP.POP.TOTL']

# Sort data for 2021

total_population_sorted = total_population_data.sort_values(by="2021", ascending=False)

top_ten_countries=total_population_sorted.head(10)
print( "Top ten countries \n")
print(top_ten_countries[['Country Code']])
```

```python
# create the BAR Plot

#top ten countries in 2021 and 2019

plt.figure(figsize=(15,9))
sns.barplot(x="2021",y="Country Code",data=top_ten_countries, palette="coolwarm")
plt.title("Top ten countries of total population 2021", fontsize=11)
plt.xlabel("Total Population",fontsize=11)
plt.ylabel("Country", fontsize=11)
plt.show()




#top ten countries in 2021 and 2019

plt.figure(figsize=(15,9))
sns.barplot(x="2019",y="Country Code",data=top_ten_countries, palette="coolwarm")
plt.title("Top ten countries of total population 2019", fontsize=11)
plt.xlabel("Total Population",fontsize=11)
plt.ylabel("Country", fontsize=11)
plt.show()
```
